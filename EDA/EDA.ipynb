```python
# import
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## First we plot the Correlation Matrix to get a first insight

```python
df = pd.read_csv('../data/train.csv')
df = df[[
    'manufactured',
    'curb_weight',
    'power',
    'engine_cap',
    'no_of_owners',
    'depreciation', 
    'coe',
    'road_tax',
    'dereg_value',
    'mileage',
    'omv',
    'arf',
    'price'
]]

corr_matrix = df.corr()

sns.set(style='white')

fig, ax = plt.subplots(figsize=(10, 8))

sns.heatmap(corr_matrix, 
            annot=True,
            fmt=".2f",
            cmap='coolwarm',
            cbar_kws={"shrink": .8}, 
            ax=ax)

plt.title('Correlation Matrix Heatmap')

plt.show()
```

```python
df = pd.read_csv('../data/train.csv')
df = df[[
    'make',
    'manufactured',
    'depreciation', 
    'coe',
    'dereg_value',
    'mileage',
    'omv',
    'arf',
    'price'
]]

df = df[df['make'] == 'volvo']

print(len(df))
```

## Then we plot the pairwise relation graphs

```python
sns.pairplot(df)
plt.show()
```

```python
df = pd.read_csv('../data/train.csv')
df = df[(df['make'] == 'subaru') | (df['make'] == 'volvo')]

plt.figure(figsize=(6, 4))
sns.scatterplot(data=df, x='arf', y='price', hue='make', palette='viridis')

plt.xlabel('arf')
plt.ylabel('price')

plt.legend(title='make', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
